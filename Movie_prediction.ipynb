## Now lets fit the columns in perfect format 

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
    
```

```python
 movies["genres"] = movies["genres"].apply(convert)
```

```python
# Now we can see we have formated the genres column
movies.head()
```

```python
movies.iloc[0].genres
```

```python
# Lets do with keywords column now
movies["keywords"] =  movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
#Now lets check out the cast. And  from cast we take first 3 actors name

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            
            L.append(i["name"])
            counter += 1
        else:
            break
            
    return L
    
```

```python
#As you have seen that I have taken the first 3 name from the cast
movies["cast"].apply(convert3)
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
#Now from the crew part we will fetch the director name
def dire_nam(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            
            L.append(i["name"])
        
            break
    return L
    
```

```python
#Here we can see we have fetched out the director name
movies["crew"].apply(dire_nam)
```

```python
movies["crew"] = movies["crew"].apply(dire_nam)
```

```python
#Now we can see we have formatted our DataFrame
movies.head()
```

```python
#As we can see the overview section is  in string we have make it in list
movies["overview"][0]
```

```python
movies["overview"].apply(lambda x:x.split())
```

```python
movies["overview"] =  movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

## We can see that in genres , keywords , cast amd crew columns there are some words which has space between them .
## We have to make them one word other wise it will not give the accuracy to the mode

```python
# so lets to it to make in one word
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Now we can see there is no gap in one word
movies.head()
```

```python
# Now we will make another column tag and will add last four column into that
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_Data_frame = movies[["movie_id","title","tags"]]
```

```python
# Here we can see we have join in one column
new_Data_frame
```

```python
#converted tags column into string
new_Data_frame["tags"]= new_Data_frame["tags"].apply(lambda x: " ".join(x))
```
