```python
 movies = movies[["movie_id","title","overview","genres","keywords","cast", "crew"]]
```

```python
movies
```

```python
## lets check whether there are missing data or duplicate data in Frame

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# As we can see there is no missing value now
movies.isnull().sum()
```

```python
#As well no duplicated value
movies.duplicated().sum()
```

## Now lets fit the columns in perfect format 

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
    
```

```python
 movies["genres"] = movies["genres"].apply(convert)
```

```python
# Now we can see we have formated the genres column
movies.head()
```

```python
movies.iloc[0].genres
```

```python
# Lets do with keywords column now
movies["keywords"] =  movies["keywords"].apply(convert)
```
